## 📊 Plotting Electronic Band Structures

**Data file needed:** `.gnu` file (e.g., `bands.dat.gnu`)

### Quick Look
Just check if the bands look right — pass the filename, action, and Fermi energy:
```python
analyze_qe("bands.dat.gnu", "bands", e_fermi=-5.3089)
```

### Publication-Ready Plot
Limit the Y-axis to the band gap region, add high-symmetry k-point labels, and save as an image:
```python
nodes = [0.0000, 0.6667, 1.0000, 1.5774]
labels = [r'$\Gamma$', 'K', 'M', r'$\Gamma$']

analyze_qe("bands.dat.gnu", "bands", e_fermi=-5.3089, ylim=[-4, 4],
           k_nodes=nodes, k_labels=labels, save_as="publication_bands.png")
```

**Parameters:**
| Param | Meaning |
|---|---|
| `e_fermi` | Fermi energy (shifts bands to E=0) |
| `ylim` | Y-axis energy window |
| `k_nodes` | Fractional positions of high-symmetry points |
| `k_labels` | Labels for high-symmetry points (LaTeX supported) |
| `save_as` | Output image filename |

````python
from workshop_tls import analyze_qe
analyze_qe("bands.dat.gnu", "bands", e_fermi=-5.3089, xlim=[0, 1.6], ylim=[-4, 4], save_as="bands.png")
````

````python
from workshop_tools import analyze_qe
nodes = [0.0000, 0.6667, 1.0000, 1.5774]
labels = [r'$\Gamma$', 'K', 'M', r'$\Gamma$']
analyze_qe("bands.dat.gnu", "bands", e_fermi=-5.3089, xlim=[0, 1.5774], ylim=[-4, 4], k_nodes=nodes, k_labels=labels, save_as="bands.png")
````

Pass just the **prefix** (not the full filename) — the tool auto-detects all matching atomic/orbital files, sums subshells per atom, and plots atom-wise PDOS:

```python
analyze_qe("mo_s_pdos", "pdos", e_fermi=-5.1089, xlim=[-4, 4], save_as="pdos_plot.png")
```

**Parameters:**
| Param | Meaning |
|---|---|
| `e_fermi` | Fermi energy reference |
| `xlim` | Energy window to display |
| `save_as` | Output image filename |

## 🌈 Plotting Optical Properties

**Data files needed:**
1. `epsr_{prefix}.dat`
2. `epsi_{prefix}.dat`

```python
analyze_qe('crsi2n4', 'optics', c_supercell=20, t_effective=6.2,
           xlim=(0, 10), save_as='crsi2opticalgrid.png')
```

This computes and plots:
- Dielectric Function (real & imaginary)
- Absorption Coefficient
- Refractive Index
- Reflectivity

**Parameters:**
| Param | Meaning |
|---|---|
| `c_supercell` | Supercell size along the vacuum direction (Å) — used for 2D correction |
| `t_effective` | Effective slab thickness (Å) — used for 2D correction |
| `xlim` | Photon energy range (eV) to display |
| `save_as` | Output image filename |

````python
from workshop_tools import analyze_qe
analyze_qe('MoS2', 'optics', c_supercell=20, t_effective=3.12, xlim=(0, 10), save_as='crsi2opticalgrid.png')
````

## 📝 Cohesive Energy Calculator

Replace the `0.0` placeholders in the script with your calculated energies:

*   **`E_crystal =`** Enter the total energy of the full crystal unit cell.
*   **`E_Mo =`** Enter the total energy of all Molybdenum atoms in the unit cell.
*   **`E_S =`** Enter the total energy of all Sulfur atoms in the unit cell.
*   **`N =`** Enter the total number of atoms (currently set to 3).

````python
Enter your calculated energies here
E_crystal = -178.24153452 
E_Mo = -136.69890473      
E_S = -20.21593269       
N = 3           # Total number of atoms

# calculation
E_total = (E_crystal - E_Mo - (E_S*2))*13.6057
print(f"Total Cohesive Energy: {E_total}")
print(f"Cohesive Energy Per Atom: {E_total / N}")
````

## 📝 HER Activity

Replace the `0.0` placeholders in the script with your calculated energies:

*   **`E_slab_H =`** Enter the total energy of the slab with the adsorbed Hydrogen.
*   **`E_slab =`** Enter the total energy of the bare slab (without Hydrogen).
*   **`E_H2 =`** Enter the total energy of an isolated H2 gas molecule.
*   **`ZPE =`** Enter the Zero Point Energy correction term.
*   **`pH =`** Enter the pH value for your system.

````python
#Enter your calculated values here
E_slab_H = -1605.21334198  
E_slab = -1604.19698179    
E_H2 = -2.33217364      
ZPE = 0.0       
pH = 0.0        

# Script calculates and prints the output automatically
Delta_E = (E_slab_H - E_slab - (0.5 * E_H2))*13.6057
Delta_G = Delta_E + ZPE - pH

print(f"Adsorption Energy (Delta E): {Delta_E}")
print(f"Gibbs Free Energy / HER Activity (Delta G_H): {Delta_G}")
````
