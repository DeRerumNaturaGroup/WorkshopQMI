## Baseline model

```python
import torch
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")


def baseline_model(nodes, w, b):
    """
    nodes: (n_atoms, n_elements)
    w:     (n_elements,)
    b:     scalar
    """
    # get sum of each element type
    atom_count = torch.sum(nodes, dim=0)
    yhat = atom_count @ w + b
    return yhat


def baseline_loss(nodes, y, w, b):
    return (baseline_model(nodes, w, b) - y) ** 2


# Parameters
w = torch.ones(16, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

# Example
nodes = torch.randn(5, 16)
y = torch.tensor(3.5)

loss = baseline_loss(nodes, y, w, b)

# Compute gradients
loss.backward()

print("Loss:", loss.item())
print("dw:", w.grad)
print("db:", b.grad)
```
