## Variables
Variables are alpha-numeric names that you can attach to data of different types, using assignment statements, to refer to them later in code. These names must start with an alphabet or an underscore. Python support many types of data out-of-the-box.

````python
# Participant's data
total_expense = 32000  # integer data
num_particpants = 534

participant1 = "Taj"   # string data
discount_rate = 0.15   # float data
is_eligible = True     # boolean data


print(f"Total expense: {total_expense}")
print(f"Number of participants: {num_particpants}")
print(f"Participant 1: {participant1}")
print(f"Discount rate: {discount_rate}")
print(f"Is participant1 eligible for discount?: {is_eligible}")
````

## Operations
Python supports the usual arithmetic operators: + (addition), * (multiplication), / (division), ** (power), // (integer division).

````python
fee_per_participant = total_expense/num_particpants
print(f"Actual fee per participant: {fee_per_participant}")

rounded_fee = total_expense//num_particpants
print(f"Every particpant will pay: {rounded_fee}")

uncovered_expense = total_expense - (rounded_fee * num_particpants)
print(f"We'll have to cover {uncovered_expense} rupees ourselves.")
````

## Conditionals
Conditional statements can skip parts of codes based on a condition. The syntax is like:

```
if conditon:
  code to be executed if condition is true
else: # optional
  code to be executed if condition is not true
```

*condition* part must result in a boolean value, i.e., *true* or *false*.


````python
discounted_fee = rounded_fee - (rounded_fee * discount_rate)
print(f"Discounted fee per participant: {discounted_fee}")

p1 = "Taj"
p2 = "Asad"
p1_age = 32
p2_age = 29

if(p1_age < 30):
  print(f"{p1} is eligible for discount; will pay {discounted_fee}.")
else:
  print(f"{p1} is not eligible for discount; will pay {rounded_fee}.")


if(p2_age < 30):
  print(f"{p2} is eligible for discount; will pay {discounted_fee}.")
else:
  print(f"{p2} is not eligible for discount; will pay {rounded_fee}.")
````

## Conditionals contd.
Conditions can be formed by uisng comparison operators: == (equals to), !=(not equal to), < (less than), <= (less than or equals to), > (greater than), >= (greater than or equals to), etc.

Further you can use logical operators to combine comparison operators inorder to make complex conditions: and (both conditions have to be true), or (any condition has to be true), etc.

The generalized syntax for conditional statements is:
```
if conditon:
  code to be executed if condition is true
elif condition2: # [optional]
  code to be executed if condition2 is true
elif condition3: # [optional]
  code to be executed if condition3 is true
...
else: # [optional]
  code to be executed if none of the above conditions is true
```

In an *if/elif/else* statement, at any given run, only one of the 'branches' will be executed. Change the value of *temp* multple times and run the cell below.

````python
temp = 52.5

if (temp <= 10):
  print("It's cold.")
elif (temp>10 and temp <= 25):
  print("It's pleasant :)")
elif (temp>=25 and temp <= 35):
  print("It's getting warm.")
elif (temp>=35 and temp<=40):
  print("It's hot.")
else:
  print("It's too hot :(")
````

# Loops
Loops can be used to run parts of code repeatedly. The most common one is the 'for loop'. Its syntax is:

```
for var in range:
  run this code repeatedly
```

````python
# the range() 'function' generates a series of values in a given range
for i in range(5):
  print(f"The value of i in this iteration is: {i}")
````

````python
# the generic form of range function is: range(start, end, step)
for i in range(1,10,2):
  print(f"The value of i in this iteration is: {i}")
````

````python
# one can combine different types of statements together
for num in range(2, 20):
  is_prime = True # initially assume prime
  for i in range(2, num):
    if (num % i == 0):
      is_prime = False
  if(is_prime):
    print(f"{num} is a prime number.")
````

# Lists
Python provides you with structures to store and manipulate collections of data. The most common such structure is a list which is a modifiable collection of data. Python uses square brackets to denote a list:


````python
# make a new list
mylist = [3.2, 44.6, "str1", 32, True, "str2"]

# the builtin function len() can be used obtain size of a collection
print(f"mylist contains {len(mylist)} elements")

# each individual element of a list can be accessed using an index,
# starting from 0 for the first element:
print(f"First element: {mylist[0]}")
print(f"Second element: {mylist[1]}")
print(f"Third element: {mylist[2]}")
print(f"Last element: {mylist[5]}")
print(f"Last element: {mylist[-1]}")  # negative indices for reverse indexing
````

````python
# collections and indexing pair well with loops
for i in range(len(mylist)):
  print(f"Element at index {i} is {mylist[i]}")
````

````python
# one can add and remove elements from list.
newlist = []
for i in range(10):
  newlist.append(i) # the dot (.) operator applies on the variable before it
print(newlist)

# remove all even elements
for i in range(len(newlist)-1, -1, -1):
  if (newlist[i] % 2 == 0):
    newlist.remove(newlist[i])
print(newlist)

# to remove at specific index, use pop function
newlist.pop(0)
print(newlist)
````

## Dictionaries
Dictionaries are collections of *(key, value)* pairs. We can use the key, instead of an index to access a particular value. Python uses the curly braces to define a dictionary and colon for key-value separation within a pair.

````python
# define a  new dictionary
ages = {"taj":32, "asad":29, "sherry":25, "rao":16, "hamza":18}

# values can be accessed using keys
print(f"rao is {ages['rao']} years old.")
print(f"sherry is {ages['sherry']} years old.")

print("----------")
# loop over pairs
for pair in ages.items():
  print(f"{pair[0]} is {pair[1]} years old.")

print("----------")
# loop over keys
for key in ages.keys():
  print(f"{key} is {ages[key]} years old.")

print("----------")
# one can also de-structure the pair when accessing it
for name, age in ages.items():
  print(f"{name} is {age} years old.")
````

## Functions
Function are blocks of codes that can be run repeatedly by passing them input parameters. Like, mathematical functions, they return a value based on the inputs passed.

Functions are 'defined' first and then 'called' wherever they are needed.

````python
def add(a, b):
  sum = a + b
  return sum

v1 = 32
v2 = 55
v3 = 9
print(f"Sum of {v1} and {v2} is {add(v1, v2)}")
print(f"Sum of {v1} and {v3} is {add(v1, v3)}")

# previously we've used built-in functions 'len()', 'range()', etc.
````

````python
l1 = [1,2,3,4,5]
l2 = [2,4,6,8,10]

def dot_product(v1, v2):
  if (len(v1) != len(v2)):
    return -1

  result = 0
  for i in range(len(v1)):
    result += v1[i] * v2[i]
  return result

print(f"The dot product of l1 and l1 is {dot_product(l1, l1)}")
````

# Part 2: Numpy
Numpy is a Python library optimized for numerical computations on multi-dimensional arrays. Each such array has a rank (number of dimensions), a shape (a vector defining the length of each dimension), and a data type.  


Numpy is not part of defaulty Python installation so one has to install it via some package manager and then import it in your program for use. Fortunately, it's already installed in the Google Colab environment, so we simply import it.

````python
# import the numpy module renamed as np
import numpy as np

# create a 1D np-array from python list
a1 = np.array([1,2,3,4,5])
print(f"a1: {a1}")
print(f"a! rank: {a1.ndim}")
print(f"a1 shape: {a1.shape}")
print(f"a1 data type: {a1.dtype}")
print("----------")

# create a 2D np-array from python list
a2 = np.array([[1,2,3], [4,5,6]])
print(f"a2: {a2}")
print(f"a2 rank: {a2.ndim}")
print(f"a2 shape: {a2.shape}")
print(f"a2 data type: {a2.dtype}")
print("----------")

# np provides convenience functions for array creation.
# These include: np.ones(), np.zeroes(), np.full(), np.arange(), np.random.randint(), etc.
# create a 2D np-arra of all 1s.
a3 = np.ones((2,3))
print(f"a3: {a3}")
print(f"a3 rank: {a3.ndim}")
print(f"a3 shape: {a3.shape}")
print(f"a3 data type: {a3.dtype}")
print("----------")


# add a2 and a3 element wise
a4 = a2 + a3
print(f"a4: {a4}")
print(f"a4 rank: {a4.ndim}")
print(f"a4 shape: {a4.shape}")
print(f"a4 data type: {a4.dtype}")
print("----------")
````

### Accessing array parts
One can access parts of multi-dimensional arrays, i.e., from indidvidual elements, to sets of rows, planes, etc.

````python
# create a 3D array of random integers
a5 = np.random.randint(10, size=(2,3,4))
print(f"a5: {a5}")
print(f"a5 rank: {a5.ndim}")
print(f"a5 shape: {a5.shape}")
print(f"a5 data type: {a5.dtype}")
print("----------")

# to acces a single elment, one has to provie the position
# of that element along all three dimensions
print(f"a5[1,2,3]: {a5[1,2,3]}")
print("----------")

# to print the 1st plane
print(f"a5[0]: {a5[0]}")
print("----------")

# to print the 2nd row of the 1st plane
print(f"a5[0,1]: {a5[0,1]}")
print("----------")
````

### Slicing

````python
# create a 3D random array of floats
a6 = np.random.rand(2,3,4)
print(f"a6: {a6}")
print(f"a6 rank: {a6.ndim}")
print(f"a6 shape: {a6.shape}")
print(f"a6 data type: {a6.dtype}")
print("----------")

# print the first two rows of the 2nd plane
print(f"a6[1, :2]: {a6[1, :2]}")
print("----------")

# print the last 3 columns of the 1st plane
# print(f"a6[:1, :, 1:]: {a6[:1, :, -3:]}")
print(f"a6[:1, :, -3:]: {a6[:1, :, -3:]}")
print("----------")

# add the top-left 2x2 quadrant of the first plane
# with the bottom-right 2x2 quadrant of the second plane
a7 = a6[0, :2, :2] + a6[1, -2:, -2:]
print(f"a7: {a7}")
````

### Reshaping, and Broadcasting
Numpy allows to reshape and mould the array into different forms.

````python
# create a 3D array of random integers
a8 = np.random.randint(10, size=(2,3,4))
print(f"a8: {a8}")
print(f"a8 rank: {a8.ndim}")
print(f"a8 shape: {a8.shape}")
print(f"a8 data type: {a8.dtype}")
print("----------")


# reshape it to a 3x2x4 array
a9 = a8.reshape((3,2,4))
print(f"a9: {a9}")
print(f"a9 rank: {a9.ndim}")
print(f"a9 shape: {a9.shape}")
print(f"a9 data type: {a9.dtype}")
print("----------")

# reshape it to a 12x2 array
a10 = a8.reshape((12,2))
print(f"a10: {a10}")
print(f"a10 rank: {a10.ndim}")
print(f"a10 shape: {a10.shape}")
print(f"a10 data type: {a10.dtype}")
print("----------")
````

### Broadcasting
Usually, to perform operation on two different arrays, they have to be of the same dimensions. Broadcasting allows to perform operations of array of different dimensions under certain [conditions](https://numpy.org/doc/stable/user/basics.broadcasting.html).

````python
# create a 3D array of random integers
a11 = np.random.randint(10, size=(2,3,4))
print(f"a11: {a11}")
print(f"a11 rank: {a11.ndim}")
print(f"a11 shape: {a11.shape}")
print(f"a11 data type: {a11.dtype}")
print("----------")

# create a vector of 4 integers each of value 5
a12 = np.full(4, 5)
print(f"a12: {a12}")
print(f"a12 rank: {a12.ndim}")
print(f"a12 shape: {a12.shape}")
print(f"a12 data type: {a12.dtype}")
print("----------")

# add a11+a12
a13 = a11 + a12
print(f"a13: {a13}")
````

### Operations

````python
# create 2 3D arrays of integers
a14 = np.random.randint(10, size=(2,3,4))
print(f"a14: {a14}")
print("-------")
a15 = np.random.randint(10, size=(2,3,4))
print(f"a15: {a15}")
print("-------")

# add a14 and a15
a16 = a14 + a15
print(f"a16: {a16}")
print("-------")

# print sum of all numbers in a14
print(f"sum(a14): {np.sum(a14)}")
print("-------")
# print max of all numbers in a14
print(f"max(a14): {np.max(a14)}")
print("-------")
# print mean of all numbers in 14
print(f"mean(a14): {np.mean(a14)}")
print("-------")
# print standard deviation of all numbers in a14
print(f"std(a14): {np.std(a14)}")
````

#Part 3: MatplotLib

Matplotlib is one of the most popular libraries in Python ecosystem for data visualization. Like Numpy, its not installed by default but, fortunately the Google Colab has it so we can directly import it.

````python
# import numpy and matplot lib
import matplotlib.pyplot as plt
import numpy as np

xvalues = np.arange(0, 10, 0.1)
yvalues = np.random.randint(10, size=len(xvalues))
plt.plot(xvalues, yvalues)
plt.show()

print("----------------------")

# customize title, axes labels, line style, line color
y2values = np.sin(xvalues)
plt.plot(xvalues, y2values, linestyle='dashdot', color='red')
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
plt.show()


````
